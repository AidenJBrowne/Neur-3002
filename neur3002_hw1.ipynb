```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n_neurons = 5
baseline_rate = 10 / 30  # spikes per second
stimulus_start = 30  # seconds
stimulus_end = 60  # seconds
pre_stimulus_duration = 30  # seconds
post_stimulus_duration = 30  # seconds

def generate_spike_times(rate, duration):
    """Generate spike times using a Poisson process."""
    n_spikes = np.random.poisson(rate * duration)
    return np.sort(np.random.uniform(0, duration, n_spikes))

# Simulate spike times for each neuron
spike_times = []
np.random.seed(42)  # For reproducibility

for neuron_id in range(1, n_neurons + 1):
    times = []
    # Pre-stimulus period
    times.extend(generate_spike_times(baseline_rate, pre_stimulus_duration))
    
    # Stimulus period
    if neuron_id % 2 == 0:  # Even neurons (inhibited)
        stim_rate = baseline_rate / np.random.uniform(2, 3)
    else:  # Odd neurons (excited)
        stim_rate = baseline_rate * np.random.uniform(2, 3)
    times.extend(
        generate_spike_times(stim_rate, post_stimulus_duration) + stimulus_start
    )
    spike_times.append(times)

# Plot raster plot
plt.figure(figsize=(10, 6))

for neuron_idx, neuron_spikes in enumerate(spike_times):
    plt.eventplot(neuron_spikes, lineoffsets=neuron_idx + 1, colors='black')

# Add stimulus onset marker
plt.axvline(stimulus_start, color='red', linestyle='--', label='Stimulus Onset')

# Customize plot
plt.xlabel('Time (s)')
plt.ylabel('Neuron ID')
plt.title('Hypothetical Raster Plot')
plt.yticks(range(1, n_neurons + 1))
plt.legend()
plt.grid(False)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl


# Parameters
n_neurons = 5
baseline_rate = 10 / 30  # spikes per second
stimulus_start = 30  # seconds
stimulus_end = 60  # seconds
pre_stimulus_duration = 30  # seconds
post_stimulus_duration = 30  # seconds

def generate_spike_times(rate, duration):
    """Generate spike times using a Poisson process."""
    n_spikes = np.random.poisson(rate * duration)
    return np.sort(np.random.uniform(0, duration, n_spikes))

# Simulate spike times for each neuron
spike_times = []
np.random.seed(42)  # For reproducibility

for neuron_id in range(1, n_neurons + 1):
    times = []
    # Pre-stimulus period
    times.extend(generate_spike_times(baseline_rate, pre_stimulus_duration))
    
    # Stimulus period
    if neuron_id % 2 == 0:  # Even neurons (inhibited)
        stim_rate = baseline_rate / np.random.uniform(2, 3)
    else:  # Odd neurons (excited)
        stim_rate = baseline_rate * np.random.uniform(2, 3)
    times.extend(
        generate_spike_times(stim_rate, post_stimulus_duration) + stimulus_start
    )
    spike_times.append(times)

# Plot raster plot
plt.figure(figsize=(10, 6))

for neuron_idx, neuron_spikes in enumerate(spike_times):
    color = 'mediumvioletred' if (neuron_idx + 1) % 2 == 0 else 'dodgerblue'
    plt.eventplot(neuron_spikes, lineoffsets=neuron_idx + 1, colors=color)

# Add stimulus onset marker
plt.axvline(stimulus_start, color='red', linestyle='--', label='Stimulus Onset')
plt.legend(fontsize=12)
plt.axvspan(stimulus_start - 0.4, stimulus_start + 0.4, color='red', alpha=0.2,)


# Customize plot
plt.xlabel('Time (s)', fontsize=14)
plt.ylabel('Neuron Number', fontsize=14)
plt.title('Visualization of Hypothetical Neurons Firing', fontsize=16)
plt.yticks(range(1, n_neurons + 1))
plt.tick_params(axis='both', labelsize=14)
plt.legend()
plt.grid(False)
mpl.rcParams['font.family'] = 'Times New Roman'

plt.savefig('/Users/Zeynepuzun/Downloads/neur3002_hw1plot.png')

plt.show()
```
